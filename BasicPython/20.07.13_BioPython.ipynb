### 0. OT
	생명정보 분석 프로그램은 모든 회사에서 필수적으로 쓰는 것
	
	소프트웨어 환경
		리눅스 환경설정 가능, 원하는대로 조정하기 쉬워 특정목적의 프로그램을 사용할 경우 사용
		Compiler or Interpreter : 프로그래밍 언어의 종류 

	Compiler, Interpreter
		Compiler
			소스코드를 처음부터 끝까지 완벽하게 구성한 후에 실행 가능 
			먼저 프로그램을 만들고 프로그램이 똑바로 짜여져있는지 확인하기 
			똑바르게 짜여있어야만 실행이 가능 
			대신 처음부터 끝까지 어떤 일을 하는지 미리 알고 있어 빠른 길을 찾아갈 수 있음 
			프로그램 실행이 빨라짐
			메모리를 적게 먹음 그래서 퍼포먼스가 적게 먹음 
		
		Interpreter
			한줄한줄 명령을 입력하는데
			한줄한줄이 에러가 있는지 없는지 알려줌 
			파이썬은 interpreter언어
			그래서 처음부터 완벽하게 코드를 짜지 않아도 원활히 실행 진행 가능

### 1. Python 기본 사용법 

	Vim 
		텍스트 편집기
		
		기본사용법 
			vi
			i
				입력가능 
			esc 버튼
				입력 끝
			:q!
				나가기
		
		작성내용확인
			vi test.py
				vim으로 test.py라는 파일을 여는 것 기존에 없었던 파일이면 새로생긴거다
			i누르고 print("Hello World!")
			esc
			:w
				저장됨 text editing 완료 
			:q!
				나가기
			ls
				파일 생성 확인 
			cat test.py
				방금 작성한 문구를 볼 수 있음 
			python test.py
				print 내용만 제시됨
				Hello World라고 한줄 친게 python interpreter에서 한가지 명령을 시킨것
				이제 소스코드 수정, 저장, 코드실행 가능함 
			
		vimrc 설정
			vi .vimrc 
				vim 설정 
			i 누르고 밑에 입력
				# 기록 기억 개수
				set history=10000
				
				#작업 중인 파일 외부에서 변경됬을 경우 자동으로 불러옴
				set autoread

				# 자동 들여쓰기
				set autoindent

				# 타이틀바에 현재 편집중인 파일을 표시
				set title

				# statusline 보여주기
				set laststatus=2

				# 붙여넣기시 계단현상 없애기
				set paste

				# 검색어 하이라이팅
				set hlsearch

				# 키워드 입력시 점진적 검색
				set incsearch

				# 일치하는 괄호 하이라이팅
				set showmatch

				# 줄번호 
				set number

				# 현재 라인 강조
				set cursorline

				# 현재 칼럼 강조
				set cursorcolumn

				# 탭을 입력하면 공백문자로 변환
				set expandtab

				#>> , << 사용시 들여쓰기 간격
				set shiftwidth=4

				# tab 키 눌렀을 때 간격
				set softtabstop=4

				# \t 문자 간격
				set tabstop=4

				# 배경화면
				set background=dark

				set t_Co=256


				syntax enable

				# 마우스 사용 허가
				set mouse=a
				* n: Normal 모드
				* v: Visual 모드
				* i: Insert 모등
				* c: Command-line 모드
				* h: 도움말 파일에서 수정할때, 모든 모드
				* a: 모든 모드
				* r: | hint-enter | 와 | more-prompt | 프롬프트에서

				#마지막으로 수정된 곳에 커서를 위치
				au BufReadPost *
					\ if line("'\"")>0 && line("'\"") <= line("$") |
					\ exe "norm g'\"" |
					\ endif
			:w
			:q!
			
			vim .vimrc
				이거로 열면 색상 변화 먹힌것도 볼 수 있음
